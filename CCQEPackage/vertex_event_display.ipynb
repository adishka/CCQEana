```python
def get_tracks_hits_vertices( ana=None, entry=0 , debug=0 ):
    if ana is None: 
        print "ana is None! returning..."
        return
    ana.GetEntry( entry )
    tracks = ana.GetTracks()
    hits = ana.GetHits()
    vertices = ana.GetVertices()
    if debug:
        print '%d track, %d hits, %d vertices'%(int(tracks.size()),int(hits.size()),int(vertices.size()))
    for v in vertices:
        print 'vertex',v.GetVertexID(),':',v.GetTruthTopologyString()
        for plane in range(3):
            print 'plane %d (wire=%d,time=%d)'%(plane,v.GetVertexWire(plane),v.GetVertexTime(plane))
    return hits,tracks,vertices
```

```python
def draw_event_display( hits=None, tracks=None , vertices=None, do_add_legend=False, planes=[0,1,2], do_add_hit_trkKey=False ,
                       xlim=None , ylim=None ):
    fig=plt.figure(figsize=(28,8))
    for plane in planes:
        ax=fig.add_subplot(1,3,plane+1)        
        # hits
        plot_hits(hits , plane=plane , ax=ax , cmap='hot_r',do_add_hit_trkKey=do_add_hit_trkKey, tracks=tracks)
        # tracks
        plot_tracks(tracks , plane=plane , ax=ax , do_add_text=False )
        # vertices
        plot_vertices(vertices , plane=plane , ax=ax )
        # style 
        if do_add_legend and plane==2: ax.add_legend(bbox_to_anchor(1,1.05),loc=2,fontsize=20)
        if plane>0: ax.yaxis.set_major_formatter( NullFormatter() )
        if xlim is not None: ax.set_xlim(xlim[plane])
        if ylim is not None: ax.set_ylim(ylim)            
        set_axes(ax,'wire','time tick' if plane==0 else '',xticks=np.linspace(np.min(ax.get_xlim()),np.max(ax.get_xlim()),5))
```

```python
#---------------------------------------------------------------------------------------------
def plot_vertices(vertices=None , plane=0 , ax=None ):
    for v in vertices:
        
        wire,time = v.GetVertexWire(plane),v.GetVertexTime(plane)
        # draw ROI
        ROI_dimensions = [100,200]
        ax.add_patch(patches.Rectangle( (wire-ROI_dimensions[0]/2,time-ROI_dimensions[1]/2),   # (x,y)
                                       ROI_dimensions[0],          # width
                                       ROI_dimensions[1],         # height
                                       fill=False,   # remove background
                                       color='salmon'
                                      ))
        # plots hits associated with the muon and proton tracks
        muon_hits = v.GetMuonHits(plane)
        print 'muon_hits:',muon_hits
        plot_hits( muon_hits , plane=plane , ax=ax , color='black')
#         proton_hits = v.GetProtonHits(plane)
        plot_hits( v.GetProtonHits(plane) , plane=plane , ax=ax , color='blue') 
#---------------------------------------------------------------------------------------------    
```

```python
#---------------------------------------------------------------------------------------------    
def draw_vertex( hits=None, tracks=None , vertex=None, do_add_legend=False, planes=[0,1,2] ):
    fig=plt.figure(figsize=(28,8))
    for plane in planes:
        ax=fig.add_subplot(1,3,plane+1)        
        # hits
        plot_hits(hits , plane=plane , ax=ax , cmap='hot_r' )
        # tracks
        plot_tracks(tracks , plane=plane , ax=ax , do_add_text=False )
        # vertices
        plot_vertex(vertex , plane=plane , ax=ax )
        # style 
        if do_add_legend and plane==2: ax.add_legend(bbox_to_anchor(1,1.05),loc=2,fontsize=20)
        if plane>0: ax.yaxis.set_major_formatter( NullFormatter() )
        set_axes(ax,'wire','time tick' if plane==0 else '',xticks=np.linspace(np.min(ax.get_xlim()),np.max(ax.get_xlim()),5))
#---------------------------------------------------------------------------------------------            

#---------------------------------------------------------------------------------------------
def plot_vertex(vertex=None , plane=0 , ax=None ):        
    wire,time = vertex.GetVertexWire(plane),vertex.GetVertexTime(plane)
    # draw ROI        
    ROI_dimensions = [100,200]
    ax.add_patch(patches.Rectangle( (wire-ROI_dimensions[0]/2,time-ROI_dimensions[1]/2),   # (x,y)    
                                   ROI_dimensions[0],          # width                                   
                                   ROI_dimensions[1],         # height                                   
                                   fill=False,   # remove background                                   
                                   color='salmon'                                   
                                  ))
    muon_hits = vertex.GetMuonHits(plane)
    proton_hits = vertex.GetProtonHits(plane)
    ax.set_xlim(np.min(np.concatenate([[h.GetWire() for h in muon_hits],[h.GetWire() for h in proton_hits]]))-ROI_dimensions[0]
                ,np.max(np.concatenate([[h.GetWire() for h in muon_hits],[h.GetWire() for h in proton_hits]]))+ROI_dimensions[0])
    ax.set_ylim(np.min(np.concatenate([[h.GetPeakT() for h in muon_hits],[h.GetPeakT() for h in proton_hits]]))-ROI_dimensions[1]
                ,np.max(np.concatenate([[h.GetPeakT() for h in muon_hits],[h.GetPeakT() for h in proton_hits]]))+ROI_dimensions[1])
    # plots hits associated with the muon and proton tracks
    plot_hits( muon_hits , plane=plane , ax=ax , color='black')        
    plot_hits( proton_hits , plane=plane , ax=ax , color='blue') 
    
    # get RdQ around vertex for a given box size
#---------------------------------------------------------------------------------------------    
```
