## (3) MC-BNB + Cosmic DATA overlay 

```python
pairsFV, MCsamples = load_pairs_as_samples()
pur_eff,pur_eff_numbers = apply_cuts()
```

```python
reduced_OnBeam = dict()
reduced_OnBeam['no cut'] = OnBeamFV

reduced_OffBeam = dict()
reduced_OffBeam['no cut'] = OffBeamFV


# cut 1: PIDa
sam = reduced_OnBeam['no cut']
reduced_OnBeam['${PID}_A$'] = sam[sam['PIDa_assigned_proton']>8]
sam = reduced_OffBeam['no cut']
reduced_OffBeam['${PID}_A$'] = sam[sam['PIDa_assigned_proton']>8]

# cut 2: require that the longer track is the one with larger PIDa
sam = reduced_OnBeam['${PID}_A$']
reduced_OnBeam['$l_{\\mu}>l_{p}$'] = sam[sam['PIDa_long'] < sam['PIDa_short']]
sam = reduced_OffBeam['${PID}_A$']
reduced_OffBeam['$l_{\\mu}>l_{p}$'] = sam[sam['PIDa_long'] < sam['PIDa_short']]


# cut 3: |\theta_{1,2}-90^0|<60^0$
sam = reduced_OnBeam['$l_{\\mu}>l_{p}$']
reduced_OnBeam['$|\theta_{1,2}-90^0|<60^0$'] = sam[np.abs(sam['theta_12']-90)<60]
sam = reduced_OffBeam['$l_{\\mu}>l_{p}$']
reduced_OffBeam['$|\theta_{1,2}-90^0|<60^0$'] = sam[np.abs(sam['theta_12']-90)<60]

```
