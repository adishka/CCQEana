```python
pair_types   = ['1mu-1p'   ,'other pairs','cosmic' ,'CC 1p 0pi'    ]
MClabels     = ['1$\\mu$1p','other pairs','cosmic' ,'CC 1p 0$\\pi$']
MCcolors     = ['teal'     ,'red'        ,'orange' ,'blue'         ]
MCcmaps      = ['Greens'   ,'Reds'       ,'Oranges','Blues'        ]
```

# can we improve the flash-cuts?
Check by ploting the cuts-figure after all other detection cuts

```python
reduced = dict()
PIDa_p_min = 13
delta_theta_12 = 55
r_max_RdQ_CC1p0pi = 0.33
opt_box=(50,100)
box_str='[%d wires x %d ticks]'%(opt_box[0],opt_box[1])
Ru,Rv,Ry = 'RdQaroundVertex[plane 0]'+box_str,'RdQaroundVertex[plane 1]'+box_str,'RdQaroundVertex[plane 2]'+box_str

fig = plt.figure(figsize=(28,16))
bins=(np.linspace(0,4,30),np.linspace(0,400,30))
minPEcut = 0.1
maxdYZcut = 200


for i,(pair_type,label,cmap,color) in enumerate(zip(pair_types,MClabels,MCcmaps,MCcolors)):
    sam = MCbnbDATAcosmicSamples[pair_type]
    sam = sam[(sam['PIDa_assigned_proton']>PIDa_p_min)&(sam['Nflashes']>0)]
    reduced[pair_type] = sam[(sam['PIDa_long'] < sam['PIDa_short'])                             
                             &(np.abs(sam['theta_12']-90)<delta_theta_12)                             
                             &((sam[Ru]==1) | (sam[Rv]==1) | (sam[Ry]==1)                                                                           
                               | (np.sqrt( np.square(sam[Ru]-1) + np.square(sam[Rv]-1) + np.square(sam[Ry]-1) )                                
                                  <= r_max_RdQ_CC1p0pi))
                            ]
    for i_sample,sample in enumerate([sam,reduced[pair_type]]):
        ax = fig.add_subplot(2,4,i+1+4*i_sample)
        plot_2d_hist(0.001*sample['ClosestFlash_TotalPE'], sample['ClosestFlash_YZdistance'], ax=ax,bins=bins,cmap=cmap)
        set_axes(ax,r'$N_{closest-flash}^{PE}$ [kPE]',r'$d_{closest-flash}^{YZ-plane}$ [cm]' if i==0 else '',fontsize=45
                 ,xticks=[1,2,3,4],yticks=[0,100,200,300]
                )
        ax.set_title(label,fontsize=45,y=1.02,color=color)
        ax.xaxis.set_major_formatter(ticker.FormatStrFormatter('%.0f'))
        ax.yaxis.set_major_formatter(ticker.FormatStrFormatter('%.0f') if i==0 else ticker.NullFormatter())    
    
        ax.plot([minPEcut,minPEcut],[np.min(ax.get_ylim()),maxdYZcut],'-'
                ,[minPEcut,np.max(ax.get_xlim())],[maxdYZcut,maxdYZcut],'-',color='black')
        
plt.tight_layout()
```

# $PID_a$ for cosmic muons vs. muons from the neutrino interaction: this is valid only for MC/MC overlay

```python
muons_from_beam = MCbnbMCcosmicSamples['1mu-1p'][MCbnbMCcosmicSamples['1mu-1p']['pdg_long']==13]
muons_from_cosmic = MCbnbMCcosmicSamples['cosmic'][MCbnbMCcosmicSamples['cosmic']['pdg_long']==13]
```

```python
fig,ax=plt.subplots(figsize=(12,8))
bins=np.linspace(0,20,31)
for sample,color,label in zip([muons_from_beam,muons_from_cosmic]
                              ,['royalblue','tomato']
                              ,['beam-induced muons','cosmic-induced muons']):
    plt.hist(sample['PIDa_assigned_muon'],bins=bins,histtype='step',color=color,linewidth=3)
    plt.plot([0,0],[0,0],color=color,linewidth=3,label=label)
set_axes(ax,r'$PID_a^{\mu}$',do_add_legend=True,title='comparison of truth-muons from different origins')

fig=plt.figure(figsize=(16,8))
bins=(np.linspace(0.1,1.5,51),np.linspace(0,12,51))
for i,(sample,cmap,label) in enumerate(zip([muons_from_beam,muons_from_cosmic]
                                          ,['Blues','Reds']
                                          ,['beam-induced muons','cosmic-induced muons'])):
    ax = fig.add_subplot(1,2,i+1)
    plt.hist2d(sample['truth_Pmu'],sample['PIDa_assigned_muon'],bins=bins,cmap=cmap)
    set_axes(ax,r'$p_{\mu}$ [GeV/c]',r'$PID_a^{\mu}$',do_add_legend=True,title=label)
plt.tight_layout()
```
