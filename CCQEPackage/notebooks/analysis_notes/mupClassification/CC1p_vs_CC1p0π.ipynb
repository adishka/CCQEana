# $CC1p$ vs. $CC1p0\pi$
$CC1p$ = an interaction with at least 1 muon and exactly 1 proton > 200 MeV/c and no charged pions with momentum greater than 70 MeV/c

$CC1p0\pi$ = an interaction with at least 1 muon and exactly 1 proton > 200 MeV/c and no pions, and no photons or electrons outside the nucleus
