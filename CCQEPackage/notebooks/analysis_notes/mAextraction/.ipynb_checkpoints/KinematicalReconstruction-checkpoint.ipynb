# Sensitivity to the axial mass
Ok, se we know how to reconstruct the proton momentum. We now want to prove that the proton momentum is sensitive to $m_A$, and (mainly) that all we measure is FSI.
We do this by using GENIE samples produced by Afro

```python
mA_array = [0.7,0.99,1.3]
mA_names = ['0_7','0_99','1_3']
genie_samples,CC1p0pi,CC1p0pi_NoFSI = dict(),dict(),dict()
for mA,name in zip(mA_array
                   ,mA_names):
    genie_samples['mA=%.2f'%mA] = pd.read_csv('/Users/erezcohen/Desktop/uBoone/CCQEanalysis/mA/genie_files/argon40_numu_mA_'+name+'.csv')
    CC1p0pi['mA=%.2f'%mA] = genie_samples['mA=%.2f'%mA][genie_samples['mA=%.2f'%mA]['CC1p0pi']==True]
    print 'len(CC1p0pi):',len(CC1p0pi['mA=%.2f'%mA])
    
    genie_samples['NoFSI mA=%.2f'%mA] = pd.read_csv('/Users/erezcohen/Desktop/uBoone/CCQEanalysis/mA/genie_files/NoFSI_argon40_numu_mA_'+name+'.csv')
    CC1p0pi['NoFSI mA=%.2f'%mA] = genie_samples['NoFSI mA=%.2f'%mA][genie_samples['NoFSI mA=%.2f'%mA]['CC1p0pi']==True]
    CC1p0pi_NoFSI['mA=%.2f'%mA] = CC1p0pi['NoFSI mA=%.2f'%mA]
    print 'len(NoFSI CC1p0pi):',len(CC1p0pi['NoFSI mA=%.2f'%mA])
```

```python
events =  CC1p0pi['mA=%.2f'%mA]
# for col in events.columns: print col;
```

```python
fig=plt.figure(figsize=(16,12))
bins = linspace(0.1,1.75,25)
for imA,mA in enumerate(mA_array):
    
    ax=fig.add_subplot(3,2,2*imA+1)
    for FSIname,FSIlabel,color in zip(['','NoFSI ']
                                      ,['no FSI','with FSI']                              
                                      ,['royalblue','tomato']):
        ax.hist(CC1p0pi[FSIname+'mA=%.2f'%mA]['Pp'],bins=bins,histtype='step',linewidth=3,color=color,label=None);
        ax.plot([0.2,0.2],[0,0],linewidth=3,color=color,label=FSIlabel)
    set_axes(ax,r'$p^p_{gen.}$ [GeV/c]' if imA==2 else ''
             ,'counts'
             ,remove_ticks_x=False if imA==2 else True
             ,do_add_grid=True
             ,do_add_legend=True if imA==0 else False)

    
    ax=fig.add_subplot(3,2,2*imA+2)
    for FSIname,FSIlabel,color in zip(['','NoFSI ']
                                      ,['no FSI','with FSI']                              
                                      ,['royalblue','tomato']):
        h,edges = np.histogram(CC1p0pi[FSIname+'mA=%.2f'%mA]['Pp'],bins=bins,normed=1); mid=0.5*(edges[1:]+edges[:-1])
        ax.step(mid,h,linewidth=3,color=color)
        ax.plot([0.2,0.2],[0,0],linewidth=3,color=color,label=FSIlabel+' normalized')
    set_axes(ax,r'$p^p_{gen.}$ [GeV/c]' if imA==2 else ''
             ,remove_ticks_x=False if imA==2 else True,remove_ticks_y=True
             ,do_add_grid=True
             ,do_add_legend=True if imA==0 else False)

    
plt.tight_layout()
plt.subplots_adjust(hspace=0)
```

# form the ratio of low/high 
and look at it with and without FSI

```python
def get_R_vs_mA(var=None,mA_array=None,CC1p0pi=None,low_bins=None,high_bins=None):
    
    R_vs_mA = pd.DataFrame(columns=['mA','R','Rerr'])
    for i_mA,mA in enumerate(mA_array): # S.Dytman: GENIE assumes mA=0.99 GeV/c2 (Dec-30)
        sample = CC1p0pi['mA=%.2f'%mA]
        N_inbin=[]
        for vmin,vmax in zip(low_bins,high_bins):
            sample_inbin = sample[(sample[var]>vmin)&(sample[var]<vmax)]
            N_inbin.append(float(len(sample_inbin)))
        R_vs_mA = R_vs_mA.append({'mA': mA
                                  ,'R':N_inbin[0]/N_inbin[1]
                                  ,'Rerr':(N_inbin[0]/N_inbin[1])*sqrt(1./N_inbin[0]+1./N_inbin[1])
                                 }, ignore_index=True)
    return R_vs_mA
    

def observable_low_high_ratio(var='Pp',mA_array=None,figsize=(14,8)
                              ,do_add_genie_overlay=True
                              ,var_min=0.25,var_mid=0.6,var_max=1.75):

    low_bins,high_bins = np.array([var_min,var_mid]),np.array([var_mid,var_max])                    
    var_centers = 0.5*(low_bins + high_bins)

    R_vs_mA_GENIE = get_R_vs_mA(var=var,mA_array=mA_array,CC1p0pi=CC1p0pi,low_bins=low_bins,high_bins=high_bins) 
    R_vs_mA_NoFSI = get_R_vs_mA(var=var,mA_array=mA_array,CC1p0pi=CC1p0pi_NoFSI,low_bins=low_bins,high_bins=high_bins)

    fig,ax=plt.subplots(figsize=figsize)
    plt.errorbar(x=R_vs_mA_GENIE['mA'],xerr=0
                 ,y=R_vs_mA_GENIE['R'],yerr=R_vs_mA_GENIE['Rerr']
                 ,fmt='o',label='genie (Afro)',markersize=15)
    plt.errorbar(x=R_vs_mA_NoFSI['mA'],xerr=0
                 ,y=R_vs_mA_NoFSI['R'],yerr=R_vs_mA_NoFSI['Rerr'],fmt='o'
                 ,label='genie (Afro) - No FSI',markersize=15)
        
    if do_add_genie_overlay:
        N_inbin = []
        for var_min,var_max in zip(low_bins,high_bins):   
            N_inbin.append(float(len(selected_CCQElike[(selected_CCQElike['truth_'+var]>var_min)&(selected_CCQElike['truth_'+var]<var_max)])))
        GENIE_overlay_mA = [0.99]
        GENIE_overlay_R = [N_inbin[0]/N_inbin[1]]
        GENIE_overlay_Rerr = (N_inbin[0]/N_inbin[1])*sqrt(1./N_inbin[0]+1./N_inbin[1])
        plt.errorbar(x=GENIE_overlay_mA,xerr=0
                     ,y=GENIE_overlay_R,yerr=GENIE_overlay_Rerr
                     ,fmt='o',label='overlay (uncorrercted)',markersize=15)
    set_axes(ax,r'$m_A$ (GeV/c$^2$)'
             ,r'$('+var+'<%.1f)/'%var_mid+'('+var+'>%.1f)$'%var_mid+' ratio',do_add_grid=True,do_add_legend=True)
    
    
```

```python
observable_low_high_ratio(var='Pp',do_add_genie_overlay=True
                          ,mA_array=mA_array
                          ,var_min=0.0,var_max=1.2,var_mid=0.6)
```
