# data and GENIE samples

```python
def get_chi2_data_genie(h_OnBeam=None,h_OnBeam_err=None,Integral_OnBeam=1
                        ,genie=None
                        ,bins=None,chi2_xrange=None
                        ,wvar = 'MicroBooNEWeight_Pmu_theta_Pp_theta'
                        ,debug=0):
    '''
    return: 
    chi2,ndf of data-genie comparison
    '''
    # genie
    hGenie,hGenie_err = np.zeros(len(bins)-1),np.zeros(len(bins)-1)
    for i in range(len(bins)-1):
        genie_in_bin = genie[(genie[var]>bins[i])&(genie[var]<bins[i+1])]
        weights_in_bin = genie_in_bin[wvar]
        hGenie[i] = np.sum(weights_in_bin)
        hGenie_err[i] = np.sqrt(np.sum(np.square(weights_in_bin)))      
    hGenie_err = hGenie_err*(Integral_OnBeam/np.float(np.sum(hGenie)))    
    hGenie = hGenie*(Integral_OnBeam/np.float(np.sum(hGenie)))
    chi2 , ndf = chi2_two_histograms( bins=bins, chi2_xrange=chi2_xrange                                     
                                         , h1=h_OnBeam , h1err=h_OnBeam_err
                                         , h2=hGenie ,  h2err=hGenie_err, debug=debug )
    return chi2 , ndf
```

```python
nbins_array = np.arange(6,14,2)
results = dict()
for Q2min,Q2max in zip([0.0,0.05,0.1],[1.,0.9,0.7]):
    chi2_xrange=(Q2min,Q2max) 
    for nbins in nbins_array:
        res = pd.DataFrame(columns=['chi2','ndf','chi2red','mA'])
        bins=linspace(np.min(chi2_xrange),np.max(chi2_xrange),nbins)

        # beam-on
        h_OnBeam,edges = np.histogram( selected_events_tightFV['reco_Q2'] , bins=bins )
        h_OnBeam_err = np.array([np.max([1,np.sqrt(h_OnBeam[i])]) for i in range(len(h_OnBeam))])
        Integral_OnBeam = np.float(len(selected_events_tightFV))
        for mA in mA_array:
            chi2,ndf=get_chi2_data_genie(h_OnBeam=h_OnBeam,h_OnBeam_err=h_OnBeam_err,Integral_OnBeam=Integral_OnBeam
                                         ,genie=genie_CC1p0pi[mA]
                                ,bins=bins
                                ,chi2_xrange=chi2_xrange
                                ,wvar = 'MicroBooNEWeight_Pmu_theta_Pp_theta'                    
                                ,debug=0)
            res = res.append(pd.DataFrame({'chi2':chi2,'ndf':ndf
                                                                 ,'chi2red':chi2/ndf,'mA':mA}                                                          
                                                                ,index=[mA]))
        results["%d bins,%.2f<Q2<%.2f"%(nbins,chi2_xrange[0],chi2_xrange[1])] = res
print 'done.'
```

```python
fig,ax=plt.subplots(figsize=(12.944,8))

for Q2min,Q2max in zip([0.0,0.05,0.1],[1.,0.9,0.7]):
    for nbins in nbins_array:
        res = results["%d bins,%.2f<Q2<%.2f"%(nbins,Q2min,Q2max)]
        mA_min = np.argmin(res['chi2red'])
        plt.plot(res['mA'],res['chi2red'],label=('%d'%nbins+' bins'
                                                 +',$%.1f<Q^2<%.1f$'%(Q2min,Q2max)
                                                 +r' $m_A^{min}=%.2f$'%mA_min))
set_axes(ax,r'$m_A$ [GeV/c$^2$]',r'$\chi^2_{red.}$')
plt.legend(bbox_to_anchor=(1,1.05),loc=2,fontsize=25)
```
