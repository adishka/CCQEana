```python
versions['overlay date'] = '2018_08_16'
versions['Overlay'] = 'prodgenie_bnb_nu_uboone_overlay_mcc8.11_reco2'
SamplesNoFV = load_samples(date=versions['overlay date']
                              ,filename=versions['Overlay']+'_'+versions['overlay date']+'_vertices'
                              ,only_in_FV=False)
SamplesWithFV = load_samples(date=versions['overlay date']
                              ,filename=versions['Overlay']+'_'+versions['overlay date']+'_vertices'
                              ,only_in_FV=True)
```

```python
cuts_order  = ['no cut','Chi2Proton','Nflashes','MatchedFlash','length'
               ,'non-collinearity','vertex activity'
               ,'delta phi','Pt & delta phi']
cutlabels = ['preselection','$dE/dx$ profile',r'$N_{flashes}>0$','optical filter','track lengths'
             ,'non-collinearity','vertex activity'
             ,'coplanarity','$p_t<0.35$ GeV/c']
```

```python
reducedOverlayNoFV,pureffOverlayNoFV,_ = apply_cuts_to_overlay(OverlaySamples=SamplesNoFV,cuts_order=cuts_order)
pureffOverlayNoFV
```

```python
reducedOverlayWithFV,pureffOverlayWithFV,_ = apply_cuts_to_overlay(OverlaySamples=SamplesWithFV,cuts_order=cuts_order)
pureffOverlayWithFV
```

# with / without FV requirement

```python
sam = reducedOverlayWithFV['Pt & delta phi']['1mu-1p']
fig=plt.figure(figsize=(20,12))
ax=fig.add_subplot(2,3,1)
Pbins=linspace(0.1,1.3,31)
Tbins=linspace(0,np.pi,50)
Phibins=linspace(-np.pi,np.pi,50)



plot_2d_hist(sam['truth_Pmu'],sam['reco_Pmu_mcs'],bins=Pbins
             ,xlabel=r'truth '+r'$p_\mu$ [GeV/c]',ylabel=r'mcs '+r'$p_\mu$ [GeV/c]',ax=ax);
plt.plot([np.min(Pbins),np.max(Pbins)],[np.min(Pbins),np.max(Pbins)],'--',color='blue')
set_axes(ax,r'truth '+r'$p_\mu$ [GeV/c]',r'mcs '+r'$p_\mu$ [GeV/c]',title='fully contained in FV')

ax=fig.add_subplot(2,3,2)
plot_2d_hist(sam['truth_Pmu_theta'],sam['reco_Pmu_mcs_theta'],bins=Tbins
             ,xlabel=r'truth '+r'$\theta_\mu$ [rad.]',ylabel=r'pandora '+r'$\theta_\mu$ [rad.]',ax=ax);
plt.plot([np.min(Tbins),np.max(Tbins)],[np.min(Tbins),np.max(Tbins)],'--',color='blue')
set_axes(ax,r'truth '+r'$\theta_\mu$ [rad]',r'pandora '+r'$\theta_\mu$ [rad.]',title='fully contained in FV')


ax=fig.add_subplot(2,3,3)
plot_2d_hist(sam['truth_Pmu_phi'],sam['reco_Pmu_mcs_phi'],bins=Phibins
             ,xlabel=r'truth '+r'$\phi_\mu$ [rad.]',ylabel=r'pandora '+r'$\phi_\mu$ [rad.]',ax=ax);
plt.plot([np.min(Phibins),np.max(Phibins)],[np.min(Phibins),np.max(Phibins)],'--',color='blue')
set_axes(ax,r'truth '+r'$\phi_\mu$ [rad]',r'pandora '+r'$\phi_\mu$ [rad.]',title='fully contained in FV')



sam = reducedOverlayNoFV['Pt & delta phi']['1mu-1p']
ax=fig.add_subplot(2,3,4)
plot_2d_hist(sam['truth_Pmu'],sam['reco_Pmu_mcs'],bins=Pbins
             ,xlabel=r'truth '+r'$p_\mu$ [GeV/c]',ylabel=r'mcs reco. '+r'$p_\mu$ [GeV/c]',ax=ax);
plt.plot([np.min(Pbins),np.max(Pbins)],[np.min(Pbins),np.max(Pbins)],'--',color='blue')
set_axes(ax,r'truth '+r'$p_\mu$ [GeV/c]',r'mcs '+r'$p_\mu$ [GeV/c]',title='no FV requirement')



ax=fig.add_subplot(2,3,5)
plot_2d_hist(sam['truth_Pmu_theta'],sam['reco_Pmu_mcs_theta'],bins=Tbins
             ,xlabel=r'truth '+r'$\theta_\mu$ [rad.]',ylabel=r'pandora '+r'$\theta_\mu$ [rad.]',ax=ax);
plt.plot([np.min(Tbins),np.max(Tbins)],[np.min(Tbins),np.max(Tbins)],'--',color='blue')
set_axes(ax,r'truth '+r'$\theta_\mu$ [rad]',r'pandora '+r'$\theta_\mu$ [rad.]',title='no FV requirement')



ax=fig.add_subplot(2,3,6)
bins=linspace(-np.pi,np.pi,50)
plot_2d_hist(sam['truth_Pmu_phi'],sam['reco_Pmu_mcs_phi'],bins=Phibins
             ,xlabel=r'truth '+r'$\phi_\mu$ [rad.]',ylabel=r'pandora '+r'$\phi_\mu$ [rad.]',ax=ax);
plt.plot([np.min(Phibins),np.max(Phibins)],[np.min(Phibins),np.max(Phibins)],'--',color='blue')
set_axes(ax,r'truth '+r'$p_\mu$ [GeV/c]',r'mcs '+r'$p_\mu$ [GeV/c]',title='no FV requirement')



plt.tight_layout()
```
