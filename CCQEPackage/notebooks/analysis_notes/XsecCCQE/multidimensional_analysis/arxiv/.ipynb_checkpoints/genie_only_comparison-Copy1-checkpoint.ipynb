## single-differential cross-sections

```python
observable = 'Pmu'
bins=Bins[observable]; vlabel=vlabels[observable]; Vlabel=Vlabels[observable]; units=Units[observable]
xlabel=Vlabel+' ['+units+']' if units is not None else Vlabel
mid = 0.5*(bins[1:]+bins[:-1]); bin_width=(mid[1]-mid[0])
name = 'single_differential_Xsec_Pmu_%d_bins'%(len(mid))
```

# compute the three-fold cross section and integrate to get the single-differential

```python
bins1,bins2,bins3 = Bins['Pmu'], Bins['cos(theta(mu))'] , Bins['phi(mu)']
N1,N2,N3 = len(bins1)-1,len(bins2)-1,len(bins3)-1
```

```python
def compute_Xsec_in_3d(overlay_genie_CC1p = None,afro_genie_CC1p = None,NBins=5):
    for key in Limits.keys(): Bins[key] = np.linspace(Limits[key][0],Limits[key][1],NBins+1)
    global bins1,bins2,bins3,N1,N2,N3
    bins1,bins2,bins3 = Bins['Pmu'], Bins['cos(theta(mu))'] , Bins['phi(mu)']
    N1,N2,N3 = len(bins1)-1,len(bins2)-1,len(bins3)-1
        
    h = dict()
    keys = ['overlay genie','afro genie','Xsec overlay genie','Xsec overlay genie err','Xsec afro genie','Xsec afro genie err']
    for key in keys: h[key] = np.zeros((N1,N2,N3))
    N = dict()
    for i_P in range(N1):#{
        Pmin,Pmax = bins1[i_P],bins1[i_P+1]
        P_bin_width = Pmax - Pmin
        
        for i_cos_theta in range(N2):#{
            cos_theta_min,cos_theta_max = bins2[i_cos_theta],bins2[i_cos_theta+1]
            cos_theta_bin_width = cos_theta_max - cos_theta_min

            for i_phi in range(N3):#{
                phi_min,phi_max = bins3[i_phi],bins3[i_phi+1]
                phi_bin_width = phi_max - phi_min
                
                bin_width = P_bin_width * cos_theta_bin_width * phi_bin_width

                N['overlay genie'] = len(sam_in_3d_bin(overlay_genie_CC1p,
                                               'truth_Pmu',Pmin,Pmax,
                                               'truth_Pmu_cos_theta',cos_theta_min,cos_theta_max,
                                               'truth_Pmu_phi',phi_min,phi_max) )
                N['overlay genie scaled'] = N['overlay genie']*4.908e19/9.62e20
                h['overlay genie'][i_P][i_cos_theta][i_phi] = N['overlay genie']
                
                overlay_genie_Xsec_in_bin,overlay_genie_Xsec_err_in_bin = compute_Xsec(Non=N['overlay genie scaled'], 
                                                                                       Non_err = np.sqrt(N['overlay genie'])*4.908e19/9.62e20,
                                                                                       bin_width = bin_width)
                
                N['afro genie'] = len(sam_in_3d_bin(afro_genie_CC1p,
                                               'truth_Pmu',Pmin,Pmax,
                                               'truth_Pmu_cos_theta',cos_theta_min,cos_theta_max,
                                               'truth_Pmu_phi',phi_min,phi_max) )
                N['afro genie scaled'] = N['afro genie']*4.908e19/4.9e20
                h['afro genie'][i_P][i_cos_theta][i_phi] = N['afro genie']
                
                afro_genie_Xsec_in_bin,afro_genie_Xsec_err_in_bin = compute_Xsec(Non=N['afro genie scaled'], 
                                                                                       Non_err = np.sqrt(N['afro genie'])*4.908e19/4.9e20,
                                                                                       bin_width = bin_width)

                h['Xsec overlay genie'][i_P][i_cos_theta][i_phi] = overlay_genie_Xsec_in_bin
                h['Xsec overlay genie err'][i_P][i_cos_theta][i_phi] = overlay_genie_Xsec_err_in_bin
                h['Xsec afro genie'][i_P][i_cos_theta][i_phi] = afro_genie_Xsec_in_bin
                h['Xsec afro genie err'][i_P][i_cos_theta][i_phi] = afro_genie_Xsec_err_in_bin
            #} i_Pmu_phi
        #} i_Pmu_cos_theta
    #} i_Pmu
    print 'done.'    
    return h
```

```python
h = compute_Xsec_in_3d(overlay_genie_CC1p = overlay_genie_CC1p_in_limits,afro_genie_CC1p = afro_genie_CC1p_in_limits,NBins=5)
```

```python
overlay_genie_Xsec,overlay_genie_Xsec_err = np.zeros(N1),np.zeros(N1)
afro_genie_Xsec,afro_genie_Xsec_err = np.zeros(N1),np.zeros(N1)
for i_P in range(N1):#{
    overlay_genie_Xsec_err_sq_sum , afro_genie_Xsec_err_sq_sum = 0,0
    for i_cos_theta in range(N2):#{
        cos_theta_bin_width = bins2[i_cos_theta+1] - bins2[i_cos_theta]
        for i_phi in range(N3):#{
            phi_bin_width = bins3[i_phi+1] - bins3[i_phi]
            
            overlay_genie_Xsec[i_P] += h['Xsec overlay genie'][i_P][i_cos_theta][i_phi] * cos_theta_bin_width * phi_bin_width
            overlay_genie_Xsec_err_sq_sum += np.square(h['Xsec overlay genie err'][i_P][i_cos_theta][i_phi] * cos_theta_bin_width * phi_bin_width)
            afro_genie_Xsec[i_P] += h['Xsec afro genie'][i_P][i_cos_theta][i_phi] * cos_theta_bin_width * phi_bin_width
            afro_genie_Xsec_err_sq_sum += np.square(h['Xsec afro genie err'][i_P][i_cos_theta][i_phi] * cos_theta_bin_width * phi_bin_width)
        #}
    #}        
    overlay_genie_Xsec_err[i_P] = np.sqrt(overlay_genie_Xsec_err_sq_sum)
    afro_genie_Xsec_err[i_P] = np.sqrt(afro_genie_Xsec_err_sq_sum)
#}
print 'integrated overlay genie Xsec:',np.sum([overlay_genie_Xsec[i_P]*(bins1[i_P+1]-bins1[i_P]) for i_P in range(N1)]),'e-39 cm2'
```
