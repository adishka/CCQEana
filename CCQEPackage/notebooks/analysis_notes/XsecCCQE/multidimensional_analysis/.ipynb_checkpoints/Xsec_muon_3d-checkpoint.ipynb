## divide into 3 dimentions: 
$\vec{p} _{\mu}$
and separately
$\vec{p} _{p}$

```python
def sam_in_3d_bin(sam,
                  Pvar,Pmin,Pmax,
                  cos_theta_var,cos_theta_min,cos_theta_max,
                  phi_var,phi_min,phi_max):
    
    return sam[(Pmin<=sam[Pvar])&(sam[Pvar]<Pmax)
               &
               (cos_theta_min<=sam[cos_theta_var])&(sam[cos_theta_var]<cos_theta_max)
               &
               (phi_min <= 180./np.pi*sam[phi_var])&(180./np.pi*sam[phi_var] < phi_max)
              ]
```

The cross-section in 3D is computed in each bin $i$ by

$$
\left(\frac{d\sigma}{dp \cdot d\cos\theta \cdot d\phi} \right)_i 
=
\frac{N^\textrm{on}_i - N^\textrm{off}_i - B_i}{ \epsilon_i \cdot \Phi_\nu \cdot N_{\textrm{targets}} \cdot \Delta_i}
$$

where $N^\textrm{on}_i$ ($N^\textrm{off}_i$) is the number of selected events in on (off) beam in bin $i$, 
$B_i$ is the background estimated from the overlay,
$\Phi_\nu$ is the integrated neutrino flux, 
$N_{\textrm{targets}}$ the number of target scattering nuclei,
the bin width is a product of the bin widths in all three dimensions

$$
\Delta_i = \Delta p_i \Delta \cos \theta_i \Delta \phi_i
$$

and the efficiency is computed as

$$
\epsilon_{i} = \frac{{reconstructed}_i \; (\textrm{ reco-variable})}{{generated}_i \;(\textrm{ truth-variable})}
$$

where the numerator (denominator) is a function of the reconstructed (truth) variable.

```python
Limits=dict({
            'Pmu':(0.2,1.4)
            ,'cos(theta(mu))':(-0.5,0.95)
            ,'phi(mu)':(-180,180)
            ,'Pp':(0.3,1.0)
            ,'cos(theta(p))':(0.25,0.9)
            ,'phi(p)':(-180,180)
            })
NBins=7
Bins = dict()
for key in Limits.keys(): Bins[key] = np.linspace(Limits[key][0],Limits[key][1],NBins)
```

```python
def compute_Xsec_in_3d(beam_on=None,beam_off=None
                                        ,generated_CC1p=None,selected_CC1p=None,overlay=None):
    
    bins1,bins2,bins3 = Bins['Pmu'], Bins['cos(theta(mu))'] , Bins['phi(mu)']
    N1,N2,N3 = len(bins1)-1,len(bins2)-1,len(bins3)-1
    
    h = dict({'N(on)':np.zeros((N1,N2,N3)),'eff':np.zeros((N1,N2,N3)),'background':np.zeros((N1,N2,N3)),'gen. cc1p':np.zeros((N1,N2,N3)),'sel. cc1p':np.zeros((N1,N2,N3))})
    
    Xsec,Xsec_err = np.zeros((N1,N2,N3)) , np.zeros((N1,N2,N3))
    mc_Xsec,mc_Xsec_err = np.zeros((N1,N2,N3)) , np.zeros((N1,N2,N3))
    
    for i_P in range(N1):#{
        Pmin,Pmax = bins1[i_P],bins1[i_P+1]
        P_bin_width = Pmax - Pmin
        
        for i_cos_theta in range(N2):#{
            cos_theta_min,cos_theta_max = bins2[i_cos_theta],bins2[i_cos_theta+1]
            cos_theta_bin_width = cos_theta_max - cos_theta_min

            for i_phi in range(N3):#{
                phi_min,phi_max = bins3[i_phi],bins3[i_phi+1]
                phi_bin_width = phi_max - phi_min
                
                bin_width = P_bin_width * cos_theta_bin_width * phi_bin_width

                beam_on_in_bin = sam_in_3d_bin(beam_on,
                                               'reco_Pmu_mcs',Pmin,Pmax,
                                               'reco_Pmu_cos_theta',cos_theta_min,cos_theta_max,
                                               'reco_Pmu_mcs_phi',phi_min,phi_max)
                Non = len(beam_on_in_bin)
                h['N(on)'][i_P][i_cos_theta][i_phi] = Non

                Xsec_in_bin , Xsec_err_in_bin = 0 , 0
                mc_Xsec_in_bin , mc_Xsec_err_in_bin = 0 , 0
                if Non>0:#{
                    beam_off_in_bin = sam_in_3d_bin(beam_off,
                                               'reco_Pmu_mcs',Pmin,Pmax,
                                               'reco_Pmu_cos_theta',cos_theta_min,cos_theta_max,
                                               'reco_Pmu_mcs_phi',phi_min,phi_max)               
                    Noff = len(beam_off_in_bin)*OffBeam_scaling

                    generated_in_bin = sam_in_3d_bin(generated_CC1p,
                                               'truth_Pmu',Pmin,Pmax,
                                               'truth_Pmu_cos_theta',cos_theta_min,cos_theta_max,
                                               'truth_Pmu_phi',phi_min,phi_max)              
                    Ngen_CC1p = len(generated_in_bin)*Nevents['f(POT)']

                    selected_in_bin = sam_in_3d_bin(selected_CC1p,
                                               'reco_Pmu_mcs',Pmin,Pmax,
                                               'reco_Pmu_cos_theta',cos_theta_min,cos_theta_max,
                                               'reco_Pmu_mcs_phi',phi_min,phi_max) 
                    Nsel_CC1p = len(selected_in_bin)*Nevents['f(POT)']
                    
                    overlay_in_bin = sam_in_3d_bin(overlay,
                                               'reco_Pmu_mcs',Pmin,Pmax,
                                               'reco_Pmu_cos_theta',cos_theta_min,cos_theta_max,
                                               'reco_Pmu_mcs_phi',phi_min,phi_max) 
                    Novrelay = len(overlay_in_bin)*Nevents['f(POT)']
                    
                    B = Novrelay - Nsel_CC1p
                    eff , eff_err = get_eff(Ngen=Ngen_CC1p , Nsel=Nsel_CC1p)
                    
                    Xsec_in_bin,Xsec_err_in_bin = compute_Xsec(Non=Non, Noff=Noff, B=B, eff=eff, 
                                                               bin_width = bin_width,
                                                               Non_err = np.sqrt(Non),
                                                               Noff_err= np.sqrt(Noff),
                                                               B_err   = np.sqrt(Novrelay - Nsel_CC1p), 
                                                               eff_err = eff_err)
                    
                    mc_Xsec_in_bin,mc_Xsec_err_in_bin = compute_Xsec(Non=Nsel_CC1p, Noff=0, B=0, eff=eff,
                                                                     bin_width = bin_width,
                                                                     Non_err = np.sqrt(Nsel_CC1p),
                                                                     Noff_err= 0,B_err   = 0,
                                                                     eff_err = eff_err)

                    h['gen. cc1p'][i_P][i_cos_theta][i_phi] = Ngen_CC1p
                    h['sel. cc1p'][i_P][i_cos_theta][i_phi] = Nsel_CC1p
                    h['eff'][i_P][i_cos_theta][i_phi] = eff
                    h['background'][i_P][i_cos_theta][i_phi] = Noff + B
                    
                #} if len(data_in_bin) > 0
                Xsec[i_P][i_cos_theta][i_phi] = Xsec_in_bin
                Xsec_err[i_P][i_cos_theta][i_phi] = Xsec_err_in_bin
                mc_Xsec[i_P][i_cos_theta][i_phi] = mc_Xsec_in_bin
                mc_Xsec_err[i_P][i_cos_theta][i_phi] = mc_Xsec_err_in_bin
            #} i_Pmu_phi
        #} i_Pmu_cos_theta
    #} i_Pmu
    print 'done.'
    
    return h,Xsec,Xsec_err,mc_Xsec,mc_Xsec_err
```

```python
h,Xsec,Xsec_err,mc_Xsec,mc_Xsec_err = compute_Xsec_in_3d(beam_on=selected_beam_on,
                                                           beam_off=selected_beam_off,
                                                           generated_CC1p=genie_CC1p,
                                                           selected_CC1p=selected_CC1p,
                                                           overlay=selected_overlay_concat)
```

# The integrated cross section 
is computed by

$$
\sigma
= 
\sum_i {\left(\frac{d\sigma}{dp \cdot d\cos\theta \cdot d\phi} \right)_i  \Delta_i}
$$

where the bin width is a product of the bin widths in all three dimensions

$$
\Delta_i = \Delta p_i \Delta \cos \theta_i \Delta \phi_i
$$

the uncertainty is computed as


$$
\Delta \sigma
= 
\sqrt{ \sum_i  \left( {\left(\frac{d\sigma}{dp \cdot d\cos\theta \cdot d\phi} \right)_i  \Delta_i} \right)^2 }
$$



```python
# integrated cross-section
Xsec_integrated,Xsec_integrated_err_sq_sum = 0,0
mc_Xsec_integrated,mc_Xsec_integrated_err_sq_sum = 0,0


for i_P in range(len(Bins['Pmu'][:-1])):#{
    Pmin,Pmax = Bins['Pmu'][i_P],Bins['Pmu'][i_P+1]
    P_bin_width = Pmax - Pmin
    for i_cos_theta in range(len(Bins['cos(theta(mu))'][:-1])):#{
        cos_theta_min,cos_theta_max = Bins['cos(theta(mu))'][i_cos_theta],Bins['cos(theta(mu))'][i_cos_theta+1]
        cos_theta_bin_width = cos_theta_max - cos_theta_min
        for i_phi in range(len(Bins['phi(mu)'][:-1])):#{
            phi_min,phi_max = Bins['phi(mu)'][i_phi],Bins['phi(mu)'][i_phi+1]
            phi_bin_width = phi_max - phi_min
            
            bin_width = P_bin_width * cos_theta_bin_width * phi_bin_width
            Xsec_integrated += Xsec[i_P][i_cos_theta][i_phi]*bin_width
            Xsec_integrated_err_sq_sum += np.square(Xsec_err[i_P][i_cos_theta][i_phi]*bin_width)
            mc_Xsec_integrated += mc_Xsec[i_P][i_cos_theta][i_phi]*bin_width
            mc_Xsec_integrated_err_sq_sum += np.square(mc_Xsec_err[i_P][i_cos_theta][i_phi]*bin_width)
            
Xsec_integrated_err = np.sqrt(Xsec_integrated_err_sq_sum)
print 'Xsec_integrated: %.2f +/- %.2f'%(Xsec_integrated,Xsec_integrated_err),'e-39 cm2'
mc_Xsec_integrated_err = np.sqrt(mc_Xsec_integrated_err_sq_sum)
print 'mc_Xsec_integrated: %.2f +/- %.2f'%(mc_Xsec_integrated,mc_Xsec_integrated_err),'e-39 cm2'
```

# The single-differential cross section as a function of $p_\mu$ 
is computed by

$$
\left(\frac{d\sigma}{dp} \right)_j
= 
\sum_i {
\left( \frac{d}{d\cos\theta \cdot d\phi} \right)_i 
\left( \frac{d\sigma}{dp} \right)_j  \Delta \cos \theta_i \Delta \phi_i
}
$$

where the summation is only on bins of $\cos\theta$ and $\phi$

the uncertainty is computed as


$$
\Delta \left(\frac{d\sigma}{dp} \right)_j
= 
\sqrt{
\sum_i \left(
\left( \frac{d}{d\cos\theta \cdot d\phi} \right)_i 
\left( \frac{d\sigma}{dp} \right)_j  \Delta \cos \theta_i \Delta \phi_i
\right)^2
}
$$



```python
Xsec_1d,Xsec_1d_err = np.zeros(len(Bins['Pmu'])-1),np.zeros(len(Bins['Pmu'])-1)
mc_Xsec_1d,mc_Xsec_1d_err = np.zeros(len(Bins['Pmu'])-1),np.zeros(len(Bins['Pmu'])-1)

for i_P in range(len(Bins['Pmu'])-1):#{
    Xsec_1d_err_sq_sum = 0
    mc_Xsec_1d_err_sq_sum = 0

    for i_cos_theta in range(len(Bins['cos(theta(mu))'])-1):#{
        cos_theta_bin_width = Bins['cos(theta(mu))'][i_cos_theta+1] - Bins['cos(theta(mu))'][i_cos_theta]
        for i_phi in range(len(Bins['phi(mu)'])-1):#{
            phi_bin_width = Bins['phi(mu)'][i_phi+1] - Bins['phi(mu)'][i_phi]
            
            Xsec_1d[i_P] += Xsec[i_P][i_cos_theta][i_phi] * cos_theta_bin_width * phi_bin_width
            Xsec_1d_err_sq_sum += np.square(Xsec_err[i_P][i_cos_theta][i_phi] * cos_theta_bin_width * phi_bin_width)
            mc_Xsec_1d[i_P] += mc_Xsec[i_P][i_cos_theta][i_phi] * cos_theta_bin_width * phi_bin_width
            mc_Xsec_1d_err_sq_sum += np.square(mc_Xsec_err[i_P][i_cos_theta][i_phi] * cos_theta_bin_width * phi_bin_width)
        #}
    #}        
    Xsec_1d_err[i_P] = np.sqrt(Xsec_1d_err_sq_sum)
    mc_Xsec_1d_err[i_P] = np.sqrt(mc_Xsec_1d_err_sq_sum)
#}
Xsec_1d = np.array(Xsec_1d)
Xsec_1d_err = np.array(Xsec_1d_err)
mc_Xsec_1d = np.array(mc_Xsec_1d)
mc_Xsec_1d_err = np.array(mc_Xsec_1d_err)

print 'Xsec_1d:',Xsec_1d,'e-39 cm2/(GeV/c)'
```
