```python
sample = GENIEall
Q2var = 'truth_Q2'
fig,ax = plt.subplots(figsize=(16,8))


Q2_averages,Q2_stds = [],[]
N_inQ2bin,N_inQ2bin_err=[],[]
for Q2min,Q2max in zip(Q2min_bins,Q2max_bins):
    sample_inQ2bin = sample[(sample[Q2var]>Q2min)&(sample[Q2var]<Q2max)]    
    Q2_averages.append( np.mean(sample_inQ2bin[Q2var]) )
    Q2_stds.append( np.std(sample_inQ2bin[Q2var]) )
    # yield in MC
    N_inQ2bin.append( len(sample_inQ2bin) )
    N_inQ2bin_err.append( np.sqrt(len(sample_inQ2bin)) )    
N_inQ2bin , N_inQ2bin_err = normalize_yield( N_inQ2bin , N_inQ2bin_err , label='GENIEall '+Q2var) 
ratio = N_inQ2bin[0]/N_inQ2bin[1]
ratio_err = ratio*np.sqrt(np.square(N_inQ2bin_err[0]/N_inQ2bin[0]) + np.square(N_inQ2bin_err[1]/N_inQ2bin[1]))
plt.errorbar( x = Q2_averages, xerr = Q2_stds
             ,fmt='o',markersize=12,capsize=2,capthick=2             
             ,y=N_inQ2bin, yerr=N_inQ2bin_err,label='generated (GENIE), $R=%.2f\pm%.2f$    '%(ratio,ratio_err))

# calculated with mA
for i_mA,mA in enumerate([0.7 , 0.99 , 1.3]): # Steve Dytman: GENIE assumes mA=0.99, Dec-30
    Xsec_inQ2bin,Xsec_inQ2bin_err=[],[]
    for Q2min,Q2max in zip(Q2min_bins,Q2max_bins):
        sample_inQ2bin = sample[(sample[Q2var]>Q2min)&(sample[Q2var]<Q2max)]
        CCelasticXsec_array = CCelasticXsec( Q2=sample_inQ2bin['truth_Q2'], Ev=sample_inQ2bin['truth_Ev'], mA=mA )
        Xsec_inQ2bin.append( np.mean(CCelasticXsec_array) )
        Xsec_inQ2bin_err.append( 0 ) # np.std(CCelasticXsec_array)
    Xsec_inQ2bin , Xsec_inQ2bin_err = normalize_yield( Xsec_inQ2bin , Xsec_inQ2bin_err , label='calculated on truth information') 
    ratio = Xsec_inQ2bin[0]/Xsec_inQ2bin[1]
    plt.errorbar( x = Q2_averages+0.01*i_mA*np.ones(len(Q2_averages)), xerr = Q2_stds
             ,fmt='o',markersize=12,capsize=2,capthick=2,alpha=1 if mA==0.99 else 0.3
             ,y=Xsec_inQ2bin, yerr=Xsec_inQ2bin_err,label=r'calculated, $m_A=%.2f$ GeV/c$^2$, $R=%.2f$    '%(mA,ratio))
for i in range(len(Q2max_bins)): plt.plot([Q2max_bins[i],Q2max_bins[i]],ax.get_ylim(),'-',color='black',alpha=0.2)
set_axes(ax,title='calculated with truth information, before detector acceptance'
         ,x_label=r'$Q^2_{gen.}$ (GeV/c)$^2$',xlim=(0,1.25)
         ,y_label=r'yield [%]'
         ,do_add_grid=True,do_add_legend=True)
```

```python
generated = GENIE_CC1p0pi
accepted = GENIE_CC1p0pi_reconstructed

hgen,hacc,hcorr = np.zeros((2,2)),np.zeros((2,2)),np.zeros((2,2))
for i_Q2,(Q2min,Q2max) in enumerate(zip(Q2min_bins,Q2max_bins)):
    print "Q2min,Q2max:",Q2min,Q2max
    for i_Ev,(Evmin,Evmax) in enumerate(zip(Evmin_bins,Evmax_bins)):
        print "Evmin,Evmax:",Evmin,Evmax
        
        gen_inQ2Evbin = generated[(generated['truth_Q2']>Q2min)&(generated['truth_Q2']<Q2max)                                                
                                  &(generated['truth_Ev']>Evmin)&(generated['truth_Ev']<Evmax)]
        hgen[i_Q2][i_Ev] = len(gen_inQ2Evbin)
        
        acc_inQ2Evbin = accepted[(accepted['truth_Q2']>Q2min)&(accepted['truth_Q2']<Q2max)                                                                                                             
                                       &(accepted['truth_Ev']>Evmin)&(accepted['truth_Ev']<Evmax)]    
        hacc[i_Q2][i_Ev] = len(acc_inQ2Evbin)
        
        hcorr[i_Q2][i_Ev] = float(hgen[i_Q2][i_Ev])/(hrec[i_Q2][i_Ev] if hrec[i_Q2][i_Ev]>0 else 1)
hgen=hgen.T
hacc=hacc.T
hcorr=hcorr.T
```

```python
def y_yup_ydw_polynomial(x,a,a_err):
    y = [np.sum([a[i]*np.power(_x_,i) for i in range(len(a))]) for _x_ in x]
    yup = [np.sum([(a[i]+a_err[i])*np.power(_x_,i) for i in range(len(a))]) for _x_ in x]
    ydw = [np.sum([(a[i]-a_err[i])*np.power(_x_,i) for i in range(len(a))]) for _x_ in x]
    return y , yup, ydw
```

```python
def polynomial_2(x, a0,a1,a2):
    return (a0 + a1*x + a2*x*x)

def polynomial_3(x, a0,a1,a2,a3):
    return (a0 + a1*x + a2*x*x + a3*x*x*x)
```

```python
mA = 0.7
N_rand = 1000
debug = 1
Nmax_attempts = N_rand*10
acc_geant = reduced_MCbnbDATAcosmicSamples['no cut']['CC 1p 0pi']
Q2_averages,Q2_stds=[],[]
N_generated , N_generated_err = [],[] 
N_measured , N_measured_err = [],[]
N_calculated , N_calculated_err = [],[]
N_calculated_rec , N_calculated_rec_err = [],[]
h_accepted_array , h_acceptance_corrected_array , h_random_array = [], [] , []

Nmax_Q2 = len(largeQ2bins)-1
Nmax_Ev = len(largeEvbins)-1


for i,recQ2bin in zip(range(len(Q2bins)-1),Q2bins): # 
    GENIE_CC1p0pi_genQ2bin = GENIE_CC1p0pi[(GENIE_CC1p0pi['truth_Q2']>Q2bins[i]) & (GENIE_CC1p0pi['truth_Q2']<Q2bins[i+1])]

    # generated (before detector acceptance)
    # -- - - -- -- -- - -- - - - -- -- - - ---- 
    N_generated.append( len(GENIE_CC1p0pi_genQ2bin) )
    N_generated_err.append( np.sqrt( len(GENIE_CC1p0pi_genQ2bin) ) )


    acc_geant_recQ2bin = acc_geant[(acc_geant['reco_Q2']>Q2bins[i]) & (acc_geant['reco_Q2']<Q2bins[i+1])]
    acc_geant_recQ2bin = acc_geant_recQ2bin[acc_geant_recQ2bin['truth_Q2']>0]
    if debug: 
        print 'recQ2bin:',recQ2bin
        
    if debug>1:
        print 'len(acc_geant_recQ2bin):',len(acc_geant_recQ2bin)
    
    # mean Q2 value in bin
    # -- - - -- -- -- - -- - - - -- -- - - ---- 
    Q2_averages.append( np.mean(acc_geant_recQ2bin['reco_Q2']) )
    Q2_stds.append( np.std(acc_geant_recQ2bin['reco_Q2']) )

    # measured (not really measured here since its MC)
    # -- - - -- -- -- - -- - - - -- -- - - ---- 
    N_measured.append( len(acc_geant_recQ2bin) )
    N_measured_err.append( np.sqrt( len(acc_geant_recQ2bin) ) )


    # calculated  with Q2(rec.)
    # -- - - -- -- -- - -- - - - -- -- - - ---- 
    h_recEv,_ = np.histogram(acc_geant_recQ2bin['reco_Ev'],bins=largeEvbins); 
    i_rand,i_attempt=0,0
    rand_Ev_N_Xsec_array = []
    while (i_rand<N_rand and i_attempt<Nmax_attempts):
        i_attempt += 1
        rand_bin_Ev = np.random.randint(Nmax_Ev)
        bin_content = h_recEv[rand_bin_Ev]
        rand_val = np.random.uniform(np.max(h_recEv))
        if (rand_val <= bin_content):
            i_rand += 1
            rand_Ev = largeEvbins_centers[rand_bin_Ev]
            rand_Ev_N_Xsec = CCelasticXsec( Q2=Q2_averages[-1], Ev=rand_Ev, mA=mA )
            rand_Ev_N_Xsec_array.append(rand_Ev_N_Xsec)
    N_calculated_rec.append( np.mean( rand_Ev_N_Xsec_array ) )
    N_calculated_rec_err.append( np.std( rand_Ev_N_Xsec_array ) )

    
    # calculated (form the CC-elastic Xsec.)
    # -- - - -- -- -- - -- - - - -- -- - - ---- 
    h_random = np.zeros((len(largeQ2bins),len(largeEvbins)))
    
    h_accepted = fill_2d_histogram( largeQ2bins , largeEvbins , acc_geant_recQ2bin , varx='truth_Q2' , vary='truth_Ev' )
    h_accepted_array.append( h_accepted )
    h_acceptance_corrected = h_accepted * acc_corr_Q2Ev
    h_acceptance_corrected_array.append(h_acceptance_corrected)

    rand_Q2_N_Xsec_array = []
    max_rand_val = np.max(h_acceptance_corrected)
    # sample N_rand random pair of (Q2,Ev)
    i_rand=0
    i_attempt=0
    rand_Q2_array, rand_Ev_array = [] , [] 
    if debug>1:
        print 'h_acceptance_corrected.shape:',h_acceptance_corrected.shape
        print 'h_random.shape:',h_random.shape
        print '(Nmax_Q2,Nmax_Ev):',Nmax_Q2,Nmax_Ev
    while (i_rand<N_rand and i_attempt<Nmax_attempts):
        i_attempt += 1
        rand_bin_Q2 = np.random.randint(Nmax_Q2)
        rand_bin_Ev = np.random.randint(Nmax_Ev)
        bin_content = h_acceptance_corrected[rand_bin_Q2][rand_bin_Ev]
        # select a random number
        rand_val = np.random.uniform(max_rand_val)
        # check if the number is smaller or equal the bin-content at (Q2,Ev) 
        if debug>1:
            print 'Q2 bin:',i,',i_rand:',i_rand,',i_attempt:',i_attempt # if i_rand%100==0: 
            print 'h_acceptance_corrected[',rand_bin_Q2,'][',rand_bin_Ev,']:',bin_content,',rand_val:',rand_val

        if (rand_val <= bin_content):
            i_rand += 1
            h_random[rand_bin_Q2][rand_bin_Ev] += 1
            rand_Q2 = largeQ2bins_centers[rand_bin_Q2]
            rand_Ev = largeEvbins_centers[rand_bin_Ev]
            # for each point, calculate the expected cross-section
            rand_Q2_N_Xsec = CCelasticXsec( Q2=rand_Q2, Ev=rand_Ev, mA=mA )
            rand_Q2_N_Xsec_array.append(rand_Q2_N_Xsec)
            if debug>1: 
                print 'filling h_random[',rand_bin_Q2,'][',rand_bin_Ev,']'
                print "rand_Q2_N_Xsec_array[-1]:",rand_Q2_N_Xsec_array[-1]
            
    N_calculated.append( np.mean(rand_Q2_N_Xsec_array) )
    N_calculated_err.append( np.std( rand_Q2_N_Xsec_array ) )
    h_random_array.append(h_random)   
    
    print "Q2_averages[-1]:",Q2_averages[-1],",np.mean(acc_geant_recQ2bin['truth_Q2']):",np.mean(acc_geant_recQ2bin['truth_Q2'])

print 'done calculation - start plotting'
```

# The sensitivity to the axial mass
# This is the best we can do (we won't get better than this)
## So - do we have a sensitivity to the axial mass??
1. we probably have to cut out $Q^2<0.1$ (GeV/c$^2$)
2. lets see what is the slope of each line + uncertainty, and see if we can distinguish between them

```python
fig,ax = plt.subplots(figsize=(16,8))
Q2min = 0.1
for i_mA,(mA,color) in enumerate(zip([0.7,1.0,1.3],['red','forestgreen','blue'])):
    label="$m_A=%.1f$ GeV/c$^2$"%mA
    x = Q2_averages
    y = results_dict['mA=%.1f'%mA]['N_calculated']
    yerr = results_dict['mA=%.1f'%mA]['N_calculated_err']
    plt.errorbar(x=x+0.005*i_mA*np.ones(len(x)),xerr=Q2_stds
                 ,y=y,yerr=yerr,color=color
                 ,fmt='o'
                 ,markersize=10,capthick=2,capsize=2
                 ,label=None)#label="$m_A=%.1f$ GeV/c$^2$"%mA)
    
    # we fit x[1:] and y[1:] since we want only Q2>0.1, which means, here, above the first bin
    popt,cov = np.polyfit(x[1:],y[1:],1,cov=True)
    print 'cov:\n',cov
    slope,intercept = popt[0],popt[1]
    slope_err,intercept_err = np.sqrt(cov[0,0]),np.sqrt(cov[1,1])
    poly = np.poly1d(popt)
    print 'poly:',poly
    x_fit=np.linspace(Q2min,0.8,100)
    y_fit=poly(x_fit)
    label = r'%s: $%.2f(%.2f)Q^2+%.2f(%.2f)$'%(label,slope,slope_err,intercept,intercept_err)
    plt.plot(x_fit,y_fit,'-',color=color,label=label)
plt.plot([Q2min,Q2min],ax.get_ylim(),'--',color='black')
set_axes(ax,r'$Q^2$ (GeV/c)$^2$','yield [%]',title=r"calculated from $Q^2_{gen.}$",do_add_grid=True,do_add_legend=True)

# fig,ax = plt.subplots(figsize=(16,8))
# for i_mA,mA in enumerate([0.7,1.0,1.3]):
#     y=results_dict['mA=%.1f'%mA]['N_calculated_rec']
#     yerr=results_dict['mA=%.1f'%mA]['N_calculated_rec_err']
#     plt.errorbar(x=Q2_averages+0.005*i_mA*np.ones(len(Q2_averages)),xerr=Q2_stds
#                  ,y=y,yerr=yerr
#                  ,fmt='o'
#                  ,markersize=10,capthick=2,capsize=2
#                  ,label="$m_A=%.1f$ GeV/c$^2$"%mA)
# set_axes(ax,r'$Q^2$ (GeV/c)$^2$','yield [%]',title=r"calculated from $Q^2_{rec.}$",do_add_grid=True,do_add_legend=True)
```
